```python
import numpy as np
import matplotlib.pyplot as plt

M = int(input())
x = np.linspace(-10,10)

y1 = M*x*x
y2 = M*np.sin(x)

plt.plot(x,y1,color = "r", label = "Mx^2", linestyle="-")
plt.plot(x,y2, color= "b", label = "Msin(x)", linestyle="--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.title(f"Plot of and {M}x^2 and {M}sin(x)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns

data = {"Subjects": ["Maths", "Physics", "English", "Computer", "Chemistry"],
        "Score": np.random.randint(1,101,5)}
df = pd.DataFrame(data)
print(df)

colors = ["blue", "red", "green", "purple", "orange"]
sns.barplot(x = 'Subjects',
            y = 'Score',
            hue='Subjects',
            data = df,
            palette=colors,
            legend=True)

for index, row in df.iterrows():
    plt.text(index, row.Score + 1, str(row.Score), ha="center", fontsize=12, fontweight='bold')

plt.title("Student's Scores in Different Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

roll_no = int(input("Enter a number: "))
random.seed(roll_no)
arr = np.random.randn(50)
x = np.arange(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

cum_arr= np.cumsum(arr)
axes[0, 0].plot(x, cum_arr, marker='o', linestyle='-', color='b', label="Cumulative Sum")
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Data")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

noise = np.random.normal(scale=0.5, size=50)
y_noisy = arr + noise
axes[1, 1].scatter(x, y_noisy, color='r', marker='x',label="Data with Noise")
axes[1, 1].set_title("Scatter Plot with Noise")
axes[1, 1].set_xlabel("Data")
axes[1, 1].set_ylabel("Noise")
axes[1, 1].legend()
axes[1, 1].grid(True, linestyle='--', alpha=0.6)

axes[0, 1].axis('off')
axes[1, 0].axis('off')

plt.show()

```
